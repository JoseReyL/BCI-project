```python
#For each event, show every channel

for i in range(37):
    print(exp[0]['events'][3])
    plt.plot(exp[0]['data'][6][:,i])
    plt.show()
    plt.specgram(exp[0]['data'][10][:,i], Fs = 250)
    plt.show()
```

```python
#Same channel shown for every event

color = ''
n_exp = 1

for channel in range(37):
    for i in range(len(exp[n_exp]['events'])):
        if exp[n_exp]['events'][i].value[0] == 1:
            color = 'k'
        else:
            color = 'r'
        plt.plot(exp[n_exp]['data'][i][:,channel], color = color)
    plt.show()
```
